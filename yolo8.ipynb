```python
import os
import shutil
from pathlib import Path

def rename_images(directory):
    # 获取目录中所有的图片文件
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']
    image_files = [f for f in os.listdir(directory) if os.path.splitext(f)[1].lower() in image_extensions]
    
    # 对文件名进行排序，以确保重命名的顺序
    image_files.sort()
    
    # 创建一个临时目录来存储重命名的文件
    temp_dir = os.path.join(directory, 'temp_renamed')
    os.makedirs(temp_dir, exist_ok=True)
    
    try:
        # 重命名文件并移动到临时目录
        for i, filename in enumerate(image_files, start=1):
            old_path = os.path.join(directory, filename)
            new_filename = f"L{i}{Path(filename).suffix}"
            new_path = os.path.join(temp_dir, new_filename)
            
            shutil.move(old_path, new_path)
            print(f"Renamed: {filename} -> {new_filename}")
        
        # 将重命名的文件移回原目录
        for filename in os.listdir(temp_dir):
            shutil.move(os.path.join(temp_dir, filename), os.path.join(directory, filename))
        
        print(f"Renamed {len(image_files)} files.")
    
    finally:
        # 清理：删除临时目录
        shutil.rmtree(temp_dir, ignore_errors=True)

# 指定目录路径
directory = r"E:\huifeng\jpg\afterneg"

# 执行重命名操作
rename_images(directory)
```

```python
import xml.etree.ElementTree as ET
from PIL import Image, ImageDraw
import os

def parse_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    rectangles = []
    for annotation in root.findall('.//Annotation[@AnnoType="rect"]'):
        coords = []
        for coordinate in annotation.find('.//Coordinates'):
            x = float(coordinate.get('X'))
            y = float(coordinate.get('Y'))
            coords.append((x, y))
        if coords:
            # Assuming the rectangle coordinates are top-left and bottom-right
            top_left = coords[0]
            bottom_right = coords[2]
            rectangles.append((top_left[0], top_left[1], bottom_right[0], bottom_right[1]))
    return rectangles

def normalize_rectangles(rectangles):
    if not rectangles:
        return []
    
    # Find the bounding box of all rectangles
    min_x = min(rect[0] for rect in rectangles)
    min_y = min(rect[1] for rect in rectangles)
    max_x = max(rect[2] for rect in rectangles)
    max_y = max(rect[3] for rect in rectangles)
    
    # Normalize the rectangles
    width = max_x - min_x
    height = max_y - min_y
    normalized = []
    for rect in rectangles:
        norm_rect = (
            (rect[0] - min_x) / width,
            (rect[1] - min_y) / height,
            (rect[2] - min_x) / width,
            (rect[3] - min_y) / height
        )
        normalized.append(norm_rect)
    
    return normalized

def draw_rectangles_on_image(image_path, rectangles, output_path):
    with Image.open(image_path) as img:
        draw = ImageDraw.Draw(img)
        img_width, img_height = img.size
        print(f"Image dimensions: {img_width}x{img_height}")

        for i, rect in enumerate(rectangles):
            # Convert normalized coordinates to image coordinates
            left = int(rect[0] * img_width)
            top = int(rect[1] * img_height)
            right = int(rect[2] * img_width)
            bottom = int(rect[3] * img_height)

            # Enlarge the rectangle (e.g., 5 times larger)
            center_x = (left + right) // 2
            center_y = (top + bottom) // 2
            width = (right - left) * 5
            height = (bottom - top) * 5
            
            left = max(0, center_x - width // 2)
            top = max(0, center_y - height // 2)
            right = min(img_width, center_x + width // 2)
            bottom = min(img_height, center_y + height // 2)

            # Draw the rectangle
            draw.rectangle([left, top, right, bottom], outline='red', width=2)

            if (i + 1) % 100 == 0:
                print(f"Processed {i + 1} rectangles")

        print(f"Total rectangles processed: {len(rectangles)}")
        img.save(output_path)
        print(f"带标注的图像已保存到: {output_path}")

def main():
    base_path = 'E:/huifeng/svs/pos/'
    xml_file = os.path.join(base_path, '1901744002_2019-04-30 14_27_47.xml')
    image_path = os.path.join(base_path, 'output.jpg')
    output_path = os.path.join(base_path, 'annotated_output_normalized.jpg')

    if not os.path.exists(xml_file):
        print(f"XML文件不存在: {xml_file}")
        return
    if not os.path.exists(image_path):
        print(f"图像文件不存在: {image_path}")
        return

    rectangles = parse_xml(xml_file)
    print(f"从XML中解析出 {len(rectangles)} 个矩形")

    normalized_rectangles = normalize_rectangles(rectangles)
    draw_rectangles_on_image(image_path, normalized_rectangles, output_path)

if __name__ == "__main__":
    main()
```

```python
import xml.etree.ElementTree as ET
from PIL import Image, ImageDraw
import os

def parse_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    rectangles = []
    for annotation in root.findall('.//Annotation[@AnnoType="rect"]'):
        coords = []
        for coordinate in annotation.find('.//Coordinates'):
            x = float(coordinate.get('X'))
            y = float(coordinate.get('Y'))
            coords.append((x, y))
        if coords:
            # Assuming the rectangle coordinates are top-left and bottom-right
            top_left = coords[0]
            bottom_right = coords[2]
            rectangles.append((top_left[0], top_left[1], bottom_right[0], bottom_right[1]))
    return rectangles

def distribute_rectangles(rectangles, img_width, img_height):
    if not rectangles:
        return []
    
    # Find the bounding box of all rectangles
    min_x = min(rect[0] for rect in rectangles)
    min_y = min(rect[1] for rect in rectangles)
    max_x = max(rect[2] for rect in rectangles)
    max_y = max(rect[3] for rect in rectangles)
    
    # Calculate the scaling factors
    scale_x = img_width / (max_x - min_x)
    scale_y = img_height / (max_y - min_y)
    scale = min(scale_x, scale_y) * 0.9  # Use 90% of the available space
    
    distributed = []
    for rect in rectangles:
        # Scale and shift the rectangle
        new_rect = (
            int((rect[0] - min_x) * scale),
            int((rect[1] - min_y) * scale),
            int((rect[2] - min_x) * scale),
            int((rect[3] - min_y) * scale)
        )
        distributed.append(new_rect)
    
    return distributed

def draw_rectangles_on_image(image_path, rectangles, output_path):
    with Image.open(image_path) as img:
        draw = ImageDraw.Draw(img)
        img_width, img_height = img.size
        print(f"Image dimensions: {img_width}x{img_height}")

        distributed_rectangles = distribute_rectangles(rectangles, img_width, img_height)

        for i, rect in enumerate(distributed_rectangles):
            left, top, right, bottom = rect

            # Ensure left < right and top < bottom
            left, right = min(left, right), max(left, right)
            top, bottom = min(top, bottom), max(top, bottom)

            # Ensure coordinates are within image bounds
            left = max(0, min(left, img_width - 1))
            top = max(0, min(top, img_height - 1))
            right = max(0, min(right, img_width - 1))
            bottom = max(0, min(bottom, img_height - 1))

            # Draw the rectangle
            if left < right and top < bottom:
                draw.rectangle([left, top, right, bottom], outline='red', width=2)

            if (i + 1) % 100 == 0:
                print(f"Processed {i + 1} rectangles")

        print(f"Total rectangles processed: {len(rectangles)}")
        img.save(output_path)
        print(f"带标注的图像已保存到: {output_path}")

def main():
    base_path = 'E:/huifeng/svs/pos/'
    xml_file = os.path.join(base_path, '1901744002_2019-04-30 14_27_47.xml')
    image_path = os.path.join(base_path, 'output.jpg')
    output_path = os.path.join(base_path, 'annotated_output_distributed.jpg')

    if not os.path.exists(xml_file):
        print(f"XML文件不存在: {xml_file}")
        return
    if not os.path.exists(image_path):
        print(f"图像文件不存在: {image_path}")
        return

    rectangles = parse_xml(xml_file)
    print(f"推理出 {len(rectangles)} 个Positive区域")

    draw_rectangles_on_image(image_path, rectangles, output_path)

if __name__ == "__main__":
    main()
```
